```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
url_for_links = "https://market.yandex.ru/catalog/54726/list?hid=91491&glfilter=7893318%3A153043&local-offers-first=0&deliveryincluded=0&onstock=1"
```

```python
def get_html(url):
    r = requests.get(url, auth=('krainev90210', '03.489115996:10b83a7b952765b12d9ebfa9289f887d'), timeout = 180, stream = True)
    return r.text
```

```python
def get_links(url):
    soup = BeautifulSoup(get_html(url), 'lxml')
    base_information = soup.select('a[href*="reviews"]')
    links = [base_information[i]['href'] for i in range(len(base_information)) if base_information[i]['href'].startswith('/product')]
    return links
```

```python
get_links(url_for_links)
```

```python
url_for_pages = "https://market.yandex.ru/catalog/54726/list?hid=91491&glfilter=7893318%3A153061&local-offers-first=0&deliveryincluded=0&onstock=1"
```

```python
def get_total_pages(url):
    soup = BeautifulSoup(get_html(url), 'lxml')
    page_info = soup.find('div', class_ ='main').find('div', class_ ='layout layout_type_maya').find('div', class_ = 'layout layout_type_search i-bem').find('div', class_='layout__col i-bem layout__col_search-results_normal').find('div', class_='n-filter-applied-results__content preloadable').find('div', class_='n-pager i-bem')
    pages_count = re.findall('"pagesCount":....', page_info['data-bem'])
    number = int(re.findall(r'\d+', str(pages_count))[0])
    return number
```

```python
def get_catalog(url):
    soup = BeautifulSoup(get_html(url), 'lxml')
    soup_suppliment = soup.find_all('a', class_='n-filter-block__item-link')
    catalog = [soup_suppliment[i]['href'] for i in range(len(soup_suppliment))]
    return catalog
```
